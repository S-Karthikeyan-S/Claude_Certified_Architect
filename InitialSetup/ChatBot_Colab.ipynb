### Restart LiteLLM Proxy

It's essential to restart the LiteLLM proxy so it picks up the latest environment variables (your API keys) and the updated `config.yaml`. This cell will first attempt to kill any running LiteLLM process on port 8000 and then start a new one in the background.

````python
import os
import subprocess
import time
from google.colab import userdata

print("--- Restarting LiteLLM Proxy in Colab ---")

# 1. Kill any previously running LiteLLM processes on port 8000
try:
    print("Attempting to kill any existing LiteLLM processes on port 8000...")
    subprocess.run(["fuser", "-k", "8000/tcp"], capture_output=True, text=True, check=False)
    print("Existing LiteLLM processes (if any) killed.")
except FileNotFoundError:
    print("fuser command not found. Cannot kill existing processes automatically.")
except Exception as e:
    print(f"Error killing existing processes: {e}")

# 2. Get environment variables for subprocess
# Ensure all necessary API keys are passed to the subprocess
#Ensure GEMINI_API_KEY or GOOGLE_API_KEY is set for LiteLLM to use with Gemini models
google_api_key = userdata.get('googlespamkey') # Assuming a secret named 'googleapikey' for Gemini
if google_api_key:
    os.environ['GEMINI_API_KEY'] = google_api_key # LiteLLM expects GEMINI_API_KEY for Gemini
    os.environ['GOOGLE_API_KEY'] = google_api_key # Also set GOOGLE_API_KEY for broader compatibility
    print("Google API key successfully loaded from Colab secrets and set as environment variables.")
else:
    print("Warning: 'googleapikey' not found in Colab secrets. Please add it to avoid authentication errors for Gemini models.")

env_vars = os.environ.copy()

# 3. Start LiteLLM proxy server in the background
print("Starting LiteLLM proxy server on port 8000 using updated config.yaml...")
command = ["litellm", "--config", "config.yaml", "--port", "8000"]
proc = subprocess.Popen(command, preexec_fn=os.setsid, env=env_vars)

print(f"LiteLLM proxy server started as a background process with PID: {proc.pid}")

# Optional: Add a small delay to allow the server to boot up
time.sleep(5)
print("LiteLLM proxy server restarted. You can now re-run cells that interact with it.")
````

````python
def chat(messages,system=None,temperature=1.0):
    params = {
        "model":"claude-3-5-sonnet-20241022", # Your code will call this model name
        "max_tokens":1000,
        "messages":messages,
        "temperature":temperature
    }

    if system:
      params["system"] = system
    message = client.messages.create(**params)
    return message.content[0].text
````

````python
messages=[]
system = """ you are a patient tutor.
do not directly answer a student's questions.
Guide them to the solution step by step
"""
add_user_message(messages,"how to solve 5x+2=3")
answer = chat(messages,system=system,temperature=0.0)
````
